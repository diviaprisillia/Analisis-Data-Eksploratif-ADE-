# Transformasi Data

```python
import pandas as pd

df = pd.read_csv("Data mobil bekas bmw.csv")
df.drop_duplicates(inplace=True)

print("Info Dataset setelah Deteksi dan Penanganan Duplikasi:")
print(df.info())
```

Identifikasi dan Penanggulangan Nilai yang Hilang pada Variabel Harga, Transmisi, dan Bahan Bakar

```python
import pandas as pd
from sklearn.impute import SimpleImputer

df = pd.read_csv("Data mobil bekas bmw.csv")

df['price'].fillna(df['price'].median(), inplace=True)

transmission_imputer = SimpleImputer(strategy='most_frequent')
df['transmission'] = transmission_imputer.fit_transform(df[['transmission']])

fuel_type_imputer = SimpleImputer(strategy='most_frequent')
df['fuelType'] = fuel_type_imputer.fit_transform(df[['fuelType']])

print("Info Dataset setelah Penanggulangan Nilai yang Hilang:")
print(df.info())
```

Menentukan Metode Imputasi yang Paling Efektif untuk Variabel Price

```python
import pandas as pd
from sklearn.impute import SimpleImputer

df = pd.read_csv("Data mobil bekas bmw.csv")

custom_imputer = SimpleImputer(strategy='mean') 
df['price'] = custom_imputer.fit_transform(df[['price']])

print("Info Dataset setelah Menentukan Metode Imputasi:")
print(df.info())
```

Penanganan Outlier yang Efektif


```python
import pandas as pd
from scipy.stats import zscore

df = pd.read_csv("Data mobil bekas bmw.csv")
threshold = 3
z_scores = zscore(df[['engineSize', 'mileage', 'tax', 'mpg']])

outliers = (abs(z_scores) > threshold).any(axis=1)

df.loc[outliers, ['engineSize', 'mileage', 'tax', 'mpg']] = df.median()

print("Info Dataset setelah Penanganan Outlier:")
print(df.info())
```

# Distribusi Harga Mobil

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# menentukan nilai dari a dan b berdasarkan mean dan standard deviation
mean = df['price'].mean()
std_deviation = df['price'].std()
a = mean - (np.sqrt(3) * std_deviation)
b = mean + (np.sqrt(3) * std_deviation)

# Menghitung probabilitas x < 5 dan presentasenya
prob = 1 / (b - a)
prob_percent = prob * 100

print(f"Mean (rata-rata): {mean:.2f}")
print(f"Standard Deviation (deviasi standar): {std_deviation:.2f}")
print(f"a (batas bawah distribusi uniform): {a:.2f}")
print(f"b (batas atas distribusi uniform): {b:.2f}")
print(f"Probabilitas x < 5: {prob:.2f}")
print(f"Persentase probabilitas x < 5: {prob_percent:.2f}%")

# visualisasi histogram
sns.set(color_codes=True)
sns.set(rc={'figure.figsize': (10, 6)})
ax = sns.histplot(df['price'], bins=100, color='green', kde=True)
ax.set(xlabel='Data Distribution', ylabel='Frequency')

plt.show()
```

# Analisis Hubungan: Engine Size, Mileage, MPG terhadap Harga Mobil

Bagaimana kita dapat menganalisis hubungan antara variabel independen seperti Engine Size, Mileage, dan MPG dengan harga mobil untuk menentukan faktor-faktor utama yang mempengaruhi harga?  


Analisis Hubungan dengan Variabel Independen:
   - Menggunakan teknik analisis korelasi, seperti korelasi Pearson atau Spearman, untuk memahami hubungan antara variabel independen (Engine Size, Mileage, MPG) dengan harga mobil.
   - Membuat scatter plot atau pair plot untuk visualisasi hubungan antar variabel.

```python
# 1. Memuat matriks korelasi

# Analisis Hubungan antara Engine Size, Mileage, MPG, dan Harga
independent_variables = ['engineSize', 'mileage', 'mpg']
dependent_variable = 'price'

# Korelasi antara Variabel Independen dan Harga
correlation_matrix = df[independent_variables + [dependent_variable]].corr()
print('\nCorrelation Matrix:\n', correlation_matrix)
```

```python
# 2. Membuat visualisasi matriks korelasi

import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df[['engineSize', 'mileage', 'mpg', 'price']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriks Korelasi')
plt.show()
```

```python
# 3. Memuat scatterplot untuk  visualisasi hubungan antar variabel Engine Size, Mileage, MPG

# Visualisasi Scatter Plot untuk Hubungan antara Variabel Independen dan Harga
plt.figure(figsize=(15, 5))
for i, variable in enumerate(independent_variables, 1):
    plt.subplot(1, 3, i)
    sns.scatterplot(x=df[variable], y=df[dependent_variable])
    plt.title(f'Scatter Plot for {variable} vs. {dependent_variable}')
plt.tight_layout()
plt.show()
```

# Tren Harga Mobil

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import numpy as np

df = pd.read_csv('Data mobil bekas bmw.csv')
print(df.head())

# Visualisasi time series untuk melihat tren seiring waktu
plt.figure(figsize=(12, 6))
plt.scatter(df['year'], df['price'], label='Data Asli')
plt.title('Tren Harga Mobil Bekas Berdasarkan Tahun Pembuatan')
plt.xlabel('Tahun Pembuatan')
plt.ylabel('Harga Jual')

# Regresi Linier
X_linier = df[['year']].values
y_linier = df['price'].values
lin_reg = LinearRegression()
lin_reg.fit(X_linier, y_linier)
y_linier_pred = lin_reg.predict(X_linier)
plt.plot(df['year'], y_linier_pred, label='Regresi Linier', color='red')

plt.legend()
plt.show()

# Evaluasi kinerja model
mse_linier = mean_squared_error(y_linier, y_linier_pred)
print(f'Mean Squared Error (Regresi Linier): {mse_linier}')
```

# Pengujian Hipotesis

Perbedaan Harga antara Mobil dengan Transmisi Manual dan Otomatis

```python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv("Data mobil bekas bmw.csv")

manual_cars = df[df['transmission'] == 'Manual']['price']

automatic_cars = df[df['transmission'] == 'Automatic']['price']

t_stat, p_value = ttest_ind(manual_cars, automatic_cars)

print(f'Test Statistic: {t_stat}')
print(f'P-value: {p_value}')

alpha = 0.05

if p_value < alpha:
    print("Terdapat perbedaan signifikan dalam harga antara mobil dengan transmisi manual dan otomatis.")
else:
    print("Tidak terdapat perbedaan signifikan dalam harga antara mobil dengan transmisi manual dan otomatis.")
```

Perbedaan Harga antara Mobil dengan Bahan Bakar Petrol dan Diesel

```python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv("Data mobil bekas bmw.csv")

petrol_cars = df[df['fuelType'] == 'Petrol']['price']

diesel_cars = df[df['fuelType'] == 'Diesel']['price']

t_stat, p_value = ttest_ind(petrol_cars, diesel_cars)

print(f'Test Statistic: {t_stat}')
print(f'P-value: {p_value}')

alpha = 0.05

if p_value < alpha:
    print("Terdapat perbedaan signifikan dalam harga antara mobil dengan bahan bakar petrol dan diesel.")
else:
    print("Tidak terdapat perbedaan signifikan dalam harga antara mobil dengan bahan bakar petrol dan diesel.")
```
